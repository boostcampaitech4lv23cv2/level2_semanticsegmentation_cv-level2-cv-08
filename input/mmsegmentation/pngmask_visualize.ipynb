```python
import cv2
import random
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import gridspec
from matplotlib.colors import ListedColormap

from collections import defaultdict
from PIL import Image
```

```python
index = random.randint(0, 819) # train dataset   index  

fig = plt.figure(figsize=(16, 16)) 
gs = gridspec.GridSpec(nrows=1, # row   
                       ncols=2, # col   
                       width_ratios=[10, 10]
                      )

ax = []
for i in range(2):
    ax.append(plt.subplot(gs[i]))

img = mpimg.imread(f'../data/mmseg/img_dir/train_pseudo/pseudo_{index:04}.jpg') # image  
ax[0].imshow(img)
mask = mpimg.imread(f'../data/mmseg/ann_dir/train_pseudo/pseudo_{index:04}.png') # mask  
ax[1].imshow(mask)

plt.show()
```
